## Data Understanding
The yellow taxi trip records include fields capturing pick-up and drop-off dates/times, pick-up and drop-off locations, trip distances, itemized fares, rate types, payment types, and driver-reported passenger counts.

The data is stored in Parquet format (*.parquet*). The dataset is from 2009 to 2024. However, for this assignment, we will only be using the data from 2023.

The data for each month is present in a different parquet file. You will get twelve files for each of the months in 2023.

The data was collected and provided to the NYC Taxi and Limousine Commission (TLC) by technology providers like vendors and taxi hailing apps. <br>

You can find the link to the TLC trip records page here: https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page

**Trip Records**

How many rows are there? Do you think handling such a large number of rows is computationally feasible when we have to combine the data for all twelve months into one?

To handle this, we need to sample a fraction of data from each of the files. How to go about that? Think of a way to select only some portion of the data from each month's file that accurately represents the trends.

#### Sampling the Data
> One way is to take a small percentage of entries for pickup in every hour of a date. So, for all the days in a month, we can iterate through the hours and select 5% values randomly from those. Use `tpep_pickup_datetime` for this. Separate date and hour from the datetime values and then for each date, select some fraction of trips for each of the 24 hours.

To sample data, you can use the `sample()` method. Follow this syntax:

```Python
# sampled_data is an empty DF to keep appending sampled data of each hour
# hour_data is the DF of entries for an hour 'X' on a date 'Y'

sample = hour_data.sample(frac = 0.05, random_state = 42)
# sample 0.05 of the hour_data
# random_state is just a seed for sampling, you can define it yourself

sampled_data = pd.concat([sampled_data, sample]) # adding data for this hour to the DF
```

This *sampled_data* will contain 5% values selected at random from each hour.

Note that the code given above is only the part that will be used for sampling and not the complete code required for sampling and combining the data files.

Keep in mind that you sample by date AND hour, not just hour. (Why?)

Figure out how to sample and combine the files.

````python
# Sample the data
# It is recommmended to not load all the files at once to avoid memory overload
````

````python
# Create a list of file URLs to read (for each month of 2023)
base_url = 'https://raw.githubusercontent.com/mayuresh-tungare/eda-assignment/main/trip_records/'
file_list = [f'2023-{month}.parquet' for month in range(1, 13)] # Generates file names from 2023-1.parquet to 2023-12.parquet

# Initialise an empty DataFrame to store all sampled data for the entire year
df_sampled_year = pd.DataFrame()

# Iterate through the list of file names (to construct URLs) and sample one by one:
for file_name in file_list:
    file_path = base_url + file_name
    print(f"Processing {file_name}...")
    try:
        # Reading the current month's parquet file
        monthly_df = pd.read_parquet(file_path)

        # Convert 'tpep_pickup_datetime' to datetime objects
        monthly_df['tpep_pickup_datetime'] = pd.to_datetime(monthly_df['tpep_pickup_datetime'])

        # Extract date and hour for grouping to apply sampling strategy
        monthly_df['pickup_date'] = monthly_df['tpep_pickup_datetime'].dt.date
        monthly_df['pickup_hour'] = monthly_df['tpep_pickup_datetime'].dt.hour

        # DataFrame to store sampled data for the current month
        sampled_month_df = pd.DataFrame()

        # Loop through unique dates in the current month
        for date in monthly_df['pickup_date'].unique():
            daily_data = monthly_df[monthly_df['pickup_date'] == date]

            # Iterate through each hour of the selected date (0 to 23)
            for hour in range(24):
                hour_data = daily_data[daily_data['pickup_hour'] == hour]

                if not hour_data.empty:
                    # Sample 5% of the hourly data randomly to ensure representativeness
                    sample = hour_data.sample(frac=0.05, random_state=42)
                    # Add sampled data for this hour to the current month's sampled DataFrame
                    sampled_month_df = pd.concat([sampled_month_df, sample])

        # Concatenate the sampled data of the current month to the yearly dataframe
        df_sampled_year = pd.concat([df_sampled_year, sampled_month_df])
        print(f"Finished processing {file_name}. Total sampled rows so far: {len(df_sampled_year)}")

    except Exception as e:
        print(f"Error reading or processing file {file_name}: {e}")

# Assign the final sampled dataframe to 'df' and drop the temporary 'pickup_date' and 'pickup_hour' columns
df = df_sampled_year.drop(columns=['pickup_date', 'pickup_hour'])
print(f"Total sampled data for the year: {len(df)} rows.")
````

**2.3.1**  <font color = red>[10 marks]</font> <br>
Based on the above analysis, it seems that some of the outliers are present due to errors in registering the trips. Fix the outliers.

Some points you can look for:
- Entries where `trip_distance` is nearly 0 and `fare_amount` is more than 300
- Entries where `trip_distance` and `fare_amount` are 0 but the pickup and dropoff zones are different (both distance and fare should not be zero for different zones)
- Entries where `trip_distance` is more than 250  miles.
- Entries where `payment_type` is 0 (there is no payment_type 0 defined in the data dictionary)

These are just some suggestions. You can handle outliers in any way you wish, using the insights from above outlier analysis.

How will you fix each of these values? Which ones will you drop and which ones will you replace?

First, let us remove 7+ passenger counts as there are very less instances.

````python
# remove passenger_count > 6 and passenger_count = 0
sample_df = sample_df[(sample_df['passenger_count'] > 0) & (sample_df['passenger_count'] <= 6)]
print(f"Shape after removing invalid passenger counts: {sample_df.shape}")
````

````python
# Continue with outlier handling

# 1. Entries where trip_distance is nearly 0 and fare_amount is more than 300
# These are likely erroneous records as a high fare for no distance is illogical.
sample_df = sample_df[~((sample_df['trip_distance'] < 0.1) & (sample_df['fare_amount'] > 300))]
print(f"Shape after removing (small trip_distance, high fare_amount) outliers: {sample_df.shape}")

# 2. Entries where trip_distance and fare_amount are 0 but the pickup and dropoff zones are different
# A trip with 0 distance and 0 fare should have the same pickup and dropoff zone, or be removed.
sample_df = sample_df[~((sample_df['trip_distance'] == 0) & (sample_df['fare_amount'] == 0) & (sample_df['PULocationID'] != sample_df['DOLocationID']))]
print(f"Shape after removing (0 distance/fare, different zones) outliers: {sample_df.shape}")

# 3. Entries where trip_distance is more than 250 miles.
# These are extremely long distances for NYC taxis and are likely outliers.
sample_df = sample_df[sample_df['trip_distance'] <= 250]
print(f"Shape after removing very long trip_distance outliers: {sample_df.shape}")

# 4. Entries where payment_type is 0 (not defined in data dictionary)
sample_df = sample_df[sample_df['payment_type'] != 0]
print(f"Shape after removing payment_type = 0: {sample_df.shape}")

# 5. Entries where RatecodeID is 99 (not defined in data dictionary, invalid rate code)
sample_df = sample_df[sample_df['RatecodeID'] != 99]
print(f"Shape after removing RatecodeID = 99: {sample_df.shape}")
````

**3.2.1** <font color = red>[3 marks]</font> <br>
Identify slow routes by calculating the average time taken by cabs to get from one zone to another at different hours of the day.

Speed on a route *X* for hour *Y* = (*distance of the route X / average trip duration for hour Y*)

````python
sample_df['trip_duration'] = (sample_df['tpep_dropoff_datetime'] - sample_df['tpep_pickup_datetime']).dt.total_seconds() / 3600 # in hours

# Calculate speed (distance/duration), handling cases where duration is 0 to avoid division by zero
sample_df['speed_mph'] = sample_df.apply(lambda row: row['trip_distance'] / row['trip_duration'] if row['trip_duration'] > 0 else 0, axis=1)

# Extract pickup hour
sample_df['pickup_hour'] = sample_df['tpep_pickup_datetime'].dt.hour

# Group by pickup location, dropoff location, and pickup hour to find average speed
df_speed = sample_df.groupby(['PULocationID', 'DOLocationID', 'pickup_hour'])['speed_mph'].mean().reset_index()

# Sort to find slowest routes
df_speed_slowest = df_speed.sort_values(by='speed_mph', ascending=True)

print("Slowest routes by average speed:")
print(df_speed_slowest.head())
````

**3.2.5** <font color = red>[3 marks]</font> <br>
Identify top 10 zones with high hourly pickups. Do the same for hourly dropoffs. Show pickup and dropoff trends in these zones.

````python
# Find top 10 pickup and dropoff zones

# Calculate total pickups per zone per hour
hourly_pickups = sample_df.groupby(['PULocationID', 'pickup_hour']).size().reset_index(name='total_pickups')

# Calculate total dropoffs per zone per hour
hourly_dropoffs = sample_df.groupby(['DOLocationID', 'pickup_hour']).size().reset_index(name='total_dropoffs')

# Get top 10 pickup zones overall (summing across all hours)
top_10_pickup_zones_id = hourly_pickups.groupby('PULocationID')['total_pickups'].sum().nlargest(10).index

# Get top 10 dropoff zones overall (summing across all hours)
top_10_dropoff_zones_id = hourly_dropoffs.groupby('DOLocationID')['total_dropoffs'].sum().nlargest(10).index

# Merge zone names for top pickup zones
top_10_pickup_zones_data = hourly_pickups[hourly_pickups['PULocationID'].isin(top_10_pickup_zones_id)].merge(
    zones[['LocationID', 'zone']], left_on='PULocationID', right_on='LocationID', how='left'
)

# Merge zone names for top dropoff zones
top_10_dropoff_zones_data = hourly_dropoffs[hourly_dropoffs['DOLocationID'].isin(top_10_dropoff_zones_id)].merge(
    zones[['LocationID', 'zone']], left_on='DOLocationID', right_on='LocationID', how='left'
)

print("Top 10 Pickup Zones (IDs):")
print(top_10_pickup_zones_id)
print("\nTop 10 Dropoff Zones (IDs):")
print(top_10_dropoff_zones_id)

# Visualize hourly pickup trends for top 10 zones
plt.figure(figsize=(15, 8))
sns.lineplot(data=top_10_pickup_zones_data, x='pickup_hour', y='total_pickups', hue='zone', marker='o')
plt.title('Hourly Pickup Trends for Top 10 Zones')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Pickups')
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Visualize hourly dropoff trends for top 10 zones
plt.figure(figsize=(15, 8))
sns.lineplot(data=top_10_dropoff_zones_data, x='pickup_hour', y='total_dropoffs', hue='zone', marker='o')
plt.title('Hourly Dropoff Trends for Top 10 Zones')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Dropoffs')
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

**3.2.6** <font color = red>[3 marks]</font> <br>
Find the ratio of pickups and dropoffs in each zone. Display the 10 highest (pickup/drop) and 10 lowest (pickup/drop) ratios.

````python
# Calculate total pickups per zone
total_pickups_per_zone = sample_df.groupby('PULocationID').size().reset_index(name='total_pickups')
total_pickups_per_zone.rename(columns={'PULocationID': 'LocationID'}, inplace=True)

# Calculate total dropoffs per zone
total_dropoffs_per_zone = sample_df.groupby('DOLocationID').size().reset_index(name='total_dropoffs')
total_dropoffs_per_zone.rename(columns={'DOLocationID': 'LocationID'}, inplace=True)

# Merge pickups and dropoffs
pickup_dropoff_counts = pd.merge(total_pickups_per_zone, total_dropoffs_per_zone, on='LocationID', how='outer').fillna(0)

# Calculate pickup/dropoff ratio
# Handle cases where total_dropoffs is zero to avoid division by zero errors
pickup_dropoff_counts['pickup_dropoff_ratio'] = pickup_dropoff_counts.apply(
    lambda row: row['total_pickups'] / row['total_dropoffs'] if row['total_dropoffs'] > 0 else float('inf'),
    axis=1
)

# Merge with zone names for better readability
pickup_dropoff_ratios_named = pd.merge(pickup_dropoff_counts, zones[['LocationID', 'zone']], on='LocationID', how='left')

# Display top 10 highest pickup/dropoff ratios (zones with many pickups relative to dropoffs)
print("\nTop 10 Zones with Highest Pickup/Dropoff Ratios:")
print(pickup_dropoff_ratios_named.sort_values(by='pickup_dropoff_ratio', ascending=False).head(10)[['zone', 'total_pickups', 'total_dropoffs', 'pickup_dropoff_ratio']])

# Display top 10 lowest pickup/dropoff ratios (zones with many dropoffs relative to pickups)
# Exclude 'inf' values for lowest ratios as they represent zones with 0 dropoffs
print("\nTop 10 Zones with Lowest Pickup/Dropoff Ratios (excluding zones with 0 dropoffs):")
print(pickup_dropoff_ratios_named[pickup_dropoff_ratios_named['pickup_dropoff_ratio'] != float('inf')]
      .sort_values(by='pickup_dropoff_ratio', ascending=True).head(10)[['zone', 'total_pickups', 'total_dropoffs', 'pickup_dropoff_ratio']])
````

````python
# Filter for night hours (11 PM to 5 AM)
night_hours_df = sample_df[((sample_df['pickup_hour'] >= 23) | (sample_df['pickup_hour'] <= 5))]

# Identify top 10 pickup zones during night hours
night_pickup_zones = night_hours_df.groupby('PULocationID').size().nlargest(10).reset_index(name='night_pickups')
night_pickup_zones_named = night_pickup_zones.merge(zones[['LocationID', 'zone']], left_on='PULocationID', right_on='LocationID', how='left')

print("Top 10 Pickup Zones during Night Hours (11 PM - 5 AM):")
print(night_pickup_zones_named[['zone', 'night_pickups']])

# Identify top 10 dropoff zones during night hours
night_dropoff_zones = night_hours_df.groupby('DOLocationID').size().nlargest(10).reset_index(name='night_dropoffs')
night_dropoff_zones_named = night_dropoff_zones.merge(zones[['LocationID', 'zone']], left_on='DOLocationID', right_on='LocationID', how='left')

print("\nTop 10 Dropoff Zones during Night Hours (11 PM - 5 AM):")
print(night_dropoff_zones_named[['zone', 'night_dropoffs']])

# Visualise the hourly trend of top night time pickup zones
night_hourly_pickups = night_hours_df.groupby(['PULocationID', 'pickup_hour']).size().reset_index(name='night_pickups')
night_top_10_pickup_zones_data = night_hourly_pickups[night_hourly_pickups['PULocationID'].isin(night_pickup_zones_named['PULocationID'])].merge(
    zones[['LocationID', 'zone']], left_on='PULocationID', right_on='LocationID', how='left'
)

plt.figure(figsize=(15, 8))
sns.lineplot(data=night_top_10_pickup_zones_data, x='pickup_hour', y='night_pickups', hue='zone', marker='o')
plt.title('Hourly Pickup Trends for Top 10 Night-Time Zones (11 PM - 5 AM)')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Pickups')
plt.xticks([0,1,2,3,4,5,23]) # Only show relevant night hours
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Visualise the hourly trend of top night time dropoff zones
night_hourly_dropoffs = night_hours_df.groupby(['DOLocationID', 'pickup_hour']).size().reset_index(name='night_dropoffs')
night_top_10_dropoff_zones_data = night_hourly_dropoffs[night_hourly_dropoffs['DOLocationID'].isin(night_dropoff_zones_named['DOLocationID'])].merge(
    zones[['LocationID', 'zone']], left_on='DOLocationID', right_on='LocationID', how='left'
)

plt.figure(figsize=(15, 8))
sns.lineplot(data=night_top_10_dropoff_zones_data, x='pickup_hour', y='night_dropoffs', hue='zone', marker='o')
plt.title('Hourly Dropoff Trends for Top 10 Night-Time Zones (11 PM - 5 AM)')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Dropoffs')
plt.xticks([0,1,2,3,4,5,23]) # Only show relevant night hours
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

Now, let us find the revenue share for the night time hours and the day time hours. After this, we will move to deciding a pricing strategy.

**3.2.8** <font color = red>[2 marks]</font> <br>
Find the revenue share for nighttime and daytime hours.

````python
# Define nighttime hours (11 PM to 5 AM) and daytime hours (6 AM to 10 PM)
def get_time_of_day(hour):
    if (hour >= 23) or (hour <= 5):
        return 'Nighttime'
    else:
        return 'Daytime'

sample_df['time_of_day'] = sample_df['pickup_hour'].apply(get_time_of_day)

# Calculate total revenue for each time of day
revenue_by_time_of_day = sample_df.groupby('time_of_day')['total_amount'].sum().reset_index()

# Calculate the revenue share
total_revenue = revenue_by_time_of_day['total_amount'].sum()
revenue_by_time_of_day['revenue_share'] = (revenue_by_time_of_day['total_amount'] / total_revenue) * 100

print("Revenue Share by Time of Day:")
print(revenue_by_time_of_day)

# Visualize the revenue share
plt.figure(figsize=(8, 6))
sns.barplot(x='time_of_day', y='revenue_share', data=revenue_by_time_of_day, palette='pastel', hue='time_of_day', legend=False)
plt.title('Revenue Share by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Revenue Share (%)')
plt.show()
````
